File Handling

```python
#  open(), read/write modes with statement, pathlib
```

```python
#file modes:  r(read), w(write), a(append), x(create)
```

```python
#always use 'with' - autocloses the file 
```

```python
#Writing to a file
```

```python
with open("sample.txt", "w") as f:
    f.write("Line 1: Hi, how are you \n")
    f.write("Line 2: What can i do for you? \n")
print("Done writing to the file")
```

```python
#Reading to a file
```

```python
with open("sample.txt", "r") as p:
    data = p.read()
    print(data)
```

```python
#read line by line
```

```python
with open("sample.txt", 'r') as r:
    for line in r:
        print(line.strip() + "hey am an add-on")
        break
```

```python
#read line into a list
```

```python
with open("sample.txt", "r") as p:
    l = p.readlines()
    print(l)
```

```python
#append
```

```python
with open("sample.txt", "a") as p:
    p.write("Line 4: new line added")
```

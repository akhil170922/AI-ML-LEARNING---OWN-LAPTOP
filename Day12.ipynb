Decorators - imp( used in api calls in fastapi)

```python
#Decorator is a function that takes another function as an argument and returns a new function
#as an argument and returns a new function that
#extends or alters its behavior without modifying the original source code
# decorator  is define with @
```

```python
def log_decorator(func):
    def wrapper():
        print("Function starts")
        func()
        print("Function ended")
    return wrapper

@log_decorator
def process():
    print("Processing data....")
```

```python
process()
```

@Property 

```python
# double underscore '__' -> private variable
# single underscore '_' -> protected variable
```

```python
class circle:
    def __init__(self, radius):
        self._radius = radius
    @property
    def area(self):
        return 3.14 * self._radius * 2        
```

```python
c = circle(5)
```

```python
c.area    #here we dont use ()
```

```python
#area is the method, generally we use with area() like this
# but if we use property then we can access like a variable c._radius
```

```python
c._radius
```

```python
from dataclasses import dataclass

@dataclass

class circle:
    _radius : float
    @property
    def area(self):
        return 3.14 * self._radius **2
```

```python
r = circle(6)
```

```python
r.area
```

@classmethod

```python
#This method that works with a class and not with the object
```

```python
#class methods are accesible inside the class itself
```

```python
class student:
    school = "ABC school"
    def __init__(self, name):
        self.name = name
    @classmethod
    def change_school(cls, new_name):
        cls.school = new_name
```

```python
student.change_school("XYZ school")    #here the name is changed globally
```

```python
s1 = student("AKhil")
```

```python
s1.school
```

```python
student.change_school("highschool")
```

```python
p1 = student("akhil")
```

```python
p1.school = "HIGH SCHOOL"
```

```python
p1.school
```

```python
p2 = student("sangem")
```

```python
p2.school
```

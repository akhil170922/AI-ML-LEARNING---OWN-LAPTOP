class Polymorphism

```python
import math
```

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * self.radius * self.radius
class Rectangle:
    def __init__(self, width,height):
        self.width = width
        self.height = height
    def area(self):
        return self.height * self.width
class Triangle:
    def __init__(self, base, height):
        self.base = base
        self.height = height
    def area(self):
        return 0.5 * self.base * self.height
```

```python
shapes = [Circle(5), Rectangle(10, 2), Triangle(5,6)]
```

```python
for shape in shapes:
    print(f"{type(shape).__name__}: area = {shape.area():.2f}")
```

Function Polymorphism

```python
class Student:
    def display(self):
        return "student name"
class Employee:
    def display(self):
        return "employee name"
```

```python
def show_info(obj):
    return obj.display()
```

```python
show_info((Student()))
```

```python
show_info((Employee()))
```

```python
#This is the basic example of function polymorphism
#here two classes are defined with the same methods
#we call those methods with using function "show_info"
#here we pass "obj" as a parameter
```

```python
class CSVExporter:
    def export(self, data):
        return ",".join(str(x) for x in data)
class JSONExporter:
    def export(self, data):
        import json
        return json.dumps(data)
```

```python
def save_report(exporter, data):
    print(exporter.export(data))
```

```python
save_report(CSVExporter(), [1,2,3])
```

```python
save_report(JSONExporter(), [1,2,3])
```

Method Overloading

```python
class Add:
    def add(self, a, b):
        x = a + b
        return x
    def add(self, a, b, c):
        x = a + b + c
        return x
```

```python
a = Add()
```

```python
a.add(1,3,4)
```

```python
a.add(5,6)
```

```python
#Python defaultely dont support method over loading
```

```python
# To achieve method overloading we can use dispatch from multiple dispatch
```

```python
from multipledispatch import dispatch
 
```

```python
class Add_dis:

    @dispatch(int, int)

    def add(self, a, b):
        x = a + b
        return x
    
    @dispatch(int, int, int)

    def add(self, a, b, c):
        x = a+b+c
        return x
```

```python
b = Add_dis()
```

```python
b.add(1,3,4)
```

```python
b.add(1,2)
```

Inheritance - inheriting the features from parent class to child class

```python
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species
    def speak(self):
        return f"{self.name} makes a sound"
    def info(self):
        return f"{self.name} is a {self.species}"
```

```python
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, "Dog")
        self.breed = breed
    def speak(self):
        return f"{self.name} says woof!" # method overriding
    def fetch(self):
        return f"{self.name} fetches the ball"
```

```python
d = Dog("Buddy", "Labrador")
```

```python
d.breed
```

```python
d.species
```

```python
d.speak()
```

```python
d.fetch()
```

```python
d.info()
```

Multiple Inheritance

```python
class Flyable:
    def fly(self):
        return "I can Fly"
class Swimmable:
    def swim(self):
        return "I can swim"
class Duck(Animal, Flyable, Swimmable):
    def __init__(self, name):
        super().__init__(name, 'Duck')
    def speak(self):
        return f"{self.name} says Quak!"
```

```python
duck = Duck("Donald")
```

```python
duck.speak()
```

```python
duck.swim()
```

```python
duck.fly()
```

Multi Level Inheritance 

grand parents -> parents -> child

```python
class animal:
    def eat(self):
        return "Animal Eats"
class dog(animal):
    def bark(self):
        return "dog barks"
class puppy(dog):
    def weep(self):
        return "puppy weeps"
```

```python
p = puppy()
p.eat()
```

```python
p.bark()
```

```python
p.weep()
```

```python
d = dog()
d.weep()
```

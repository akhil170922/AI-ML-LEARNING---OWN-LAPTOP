Loops and Control Flow statements

if/elif/else

```python
# if condition

temp = 35
if temp>40:
    print("Extreme Heat")
elif temp>30:
    print("its hot")
elif temp>20:
    print("its pleasant")
else:
    print("its cold")
```

```python
#Nested if

age = 20
has_id = True

if age>=18:
    if has_id:
        print("You can enter")
    else:
        print("You need an id")
else:
    print("You are too young")
```

```python
x = 15
result = "Even" if x%2 == 0 else "Odd"
print(result)
```

```python
# Multiple conditions in if statements

score = 80
attendance = 90

if score>=80 and attendance>=90:
    print("You get certificate")
```

Enumerate

```python
# enumerate => index + value

fruits = ['apple', 'banana','guava','kiwi']

for idx, fruit in enumerate(fruits):
    print(f"{idx} : {fruit}")
```

zip

```python
names = ['akhil','anu','nagi','surya']
scores = [100, 90, 80, 70]
grades = ['A', 'B', 'C', 'D']

for name, score, grade in zip(names,scores,grades):
    print(f"{name}:{score}:{grade}")
```

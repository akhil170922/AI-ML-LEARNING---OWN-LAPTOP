*args and **kwargs

```python
#args are variable positional arguments
#in args, we can give multiple values with representing *args
def total(*args):
    print(args)
    return sum(args)
total(1,2,3,4,5)
```

```python
#kwargs are variable keyword arguments.
```

```python
def profile(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")
profile(name='akhil', age = '25')
```

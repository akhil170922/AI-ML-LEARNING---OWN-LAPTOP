```python
class student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
    def introduction(self):
        return f"I am {self.name} and my age is {self.age} and i got {self.grade} grade"
```

```python
s = student("akhil", 25, "O+")
```

```python
s.grade
```

```python
s.introduction()
```

```python
class Counter:
    total_count = 0
    def __init__(self, name):
        self.name = name
        self.count = 0
        Counter.total_count += 1
    def increment(self):
        self.count += 1
```

```python
c1 = Counter("page_view")
c1.increment()
```

```python
c1.count
```

```python
c1.total_count
```

```python
c2 = Counter("api_calls")
c2.increment()
```

```python
c2.count
```

```python
c2.total_count
```

```python
#def init is called every time object is created - here it is called two objects c1 c2 so it calls two times then the total count will br 2
```

Encapsulation - data hiding

```python
class DataParser:
    date_format = "%Y-%M-%D"
    def __init__(self,date_str):
        self.__date_str = date_str
    
    def display(self):
        return self.__date_str
```

```python
#Here we are hiding variable and keeping as a private variable-
#using __ before the variable like: __date_str
```

```python
date = DataParser("17-sept-2000")
```

```python
date.display()
```

```python
date.__date_str
```

```python
#here we cannot access the variable by using the object
#thats how encapsulation woks - its private
```

```python
class Employee:
    def __init__(self, salary):
        self.__salary = salary
    def get_salary(self):
        return self.__salary
    def set_salary(self, amount):
        if amount>0:
            self.__salary = amount
        else:
            print("Invalid Salary")
            
```

```python
e = Employee(5000)
e.set_salary(6000)
e.get_salary()
```

```python
e.__salary
```

```python
class Bank:
    def __init__(self, balance):
        self.__balance = balance
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Not allowed")
    def get_balance(self):
        return self.__balance
```

```python
b = Bank(1000)
```

```python
b.withdraw(50)
```

```python
b.__balance
```

```python
b.get_balance()
```

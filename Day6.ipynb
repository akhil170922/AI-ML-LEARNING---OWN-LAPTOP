String Manipulations with Comprehensions [mainly used in NLP]

```python
words = ['Akhil', 'hello', 'PYTHON']
```

```python
words
```

```python
lower_words = []
for word in words:
    lower_words.append(word.lower())
print(lower_words)
```

```python
# by using comprehension
```

```python
lower_words = [word.lower() for word in words]
lower_words
```

Flat

#Flat means -> to flatten the list or to unpack the nested list to a single list

```python
nested_list
```

```python
flat_list = []
for value in nested_list:
    for val in value:
        flat_list.append(val)
print(flat_list)
```

```python
#by using list comprehesnsion
```

```python
flat = [val for value in nested_list for val in value]
flat
```

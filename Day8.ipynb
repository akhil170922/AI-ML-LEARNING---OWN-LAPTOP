Variable Scope

```python
message = "hello"
def show_message():
    print("Insde Function: ", message)
show_message()
print("Outside Function: ", message)
```

```python
value = 10
def show_val():
    value = value + 5
    print("Inside func: ", value)
```

```python
show_val()
```

```python
value = 10
def new_val():
    global value
    value = value + 5
    print("Inside; ", value)

new_val()
print("Outside: ", value)
```

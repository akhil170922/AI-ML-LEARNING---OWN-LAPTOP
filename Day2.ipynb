String Methods

```python
# String Manupulations

# strip,lstrip,rstrip,split,substr,swapcase,capitalize,title,lower,upper,replace,join,find
```

```python
a = '   hello, world, python   '
```

```python
b = a.strip()
b
```

```python
left = a.lstrip()
left
```

```python
right = a.rstrip()
right
```

```python
b
```

```python
#split() creates a new list
```

```python
c = b.split(',')
c
```

```python
words = b.strip().split(',')
words
```

```python
sample = []
for w in words:
    sample.append(w.strip())
sample
```

```python
preprocessed = ' '.join(sample)
preprocessed
```

```python
preprocessed.upper()
```

```python
preprocessed.lower()
```

```python
preprocessed.capitalize()
```

```python
preprocessed.title()
```

```python
preprocessed.find('o')
```

```python
preprocessed.count('o')
```

```python
preprocessed.endswith('n')
```

```python
preprocessed.replace('world', 'akhil')
```
